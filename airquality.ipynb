```python
import pandas as pd
import seaborn as sns
df = pd.read_csv("airquality.csv")
```

```python
df.info()
```

```python
##volem saber la longitude del dataset que estem treballant
len(df)
```

```python
## i mostrar-ne les primeres línies
df.head()
```

```python
## ara volem conèixer la màxima temperatura enregistrada
max_temperatura = df["Temp"].max()
print(max_temperatura)
```

```python
## i volem saber aquesta temperatura en quin mes i en quin dia es va enregistrar
fila_max = df[df["Temp"] == df["Temp"].max()]
print(fila_max)
```

```python
fila_max = df[df["Temp"] == df["Temp"].max()]
print(fila_max[["Month", "Day"]])
```

```python
## veiem que fa ser el mes d'agost, concretament el dia 28.
## ens interessa saber si el vent i la radiació solar tenen relació entre elles;
## per a saber això farem un scatterplot amb cada variable en un dels dos eixos.
sns.scatterplot(data=df, x="Wind", y="Solar.R")
```

```python

import matplotlib.pyplot as plt
sns.scatterplot(data=df, x="Wind", y="Solar.R")
plt.show()
```

```python
## no es dibuixa un línia clara, de manera que podem concloure que no hi ha relació entre
## totes dues variables.
## a continuació amb un gràfic de barres mirarem quins són els mesos que fa mes calor
sns.barplot(data=df, x="Month", y="Temp")
```

```python
## és al juliol i a l'agost, cosa que ens quadra amb que és estiu, i amb que la 
## temperatura més alta registrada hem vist abans que era al mes d'agost.
## Amb el següent gràfic veiem que l'ozó i la temperatura sí que tenen relació
## entre elles. Com més creix l'ozó, més creix la temperatura
sns.scatterplot(data=df, x="Ozone", y="Temp")
```
